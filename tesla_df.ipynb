```python
# Dependencies
import pandas as pd
```

```python
file = "Resources/tesla_data.csv"
```

```python
df = pd.read_csv(file, encoding="ISO-8859-1");

df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
new_df = df.drop(['Case #','Unnamed: 16', 'Unnamed: 17', ' Source ', ' Note ', ' Deceased 1 ',
       ' Deceased 2 ', ' Deceased 3 ', ' Deceased 4 '], axis=1)
```

```python
new_df = new_df.dropna()
new_df.head()
```

```python
new_df.info()
```

```python
# Remove white spaces in column names
new_df = new_df.rename(columns=lambda x: x.strip())
```

```python
# Rename the columns
new_df.rename(columns={
    "Tesla driver": "Driver",
    "Tesla occupant": "Occupant",  
    "Autopilot claimed": "Autopilot",
    "Verified Tesla Autopilot Deaths": "Autopilot Deaths",
    "Verified Tesla Autopilot Deaths + All Deaths Reported to NHTSA SGO": "Autopilot Deaths & Reported Deaths"
})
```

```python
new_df['Country'] = new_df['Country'].str.strip()
```

```python
new_df['State'] = new_df['State'].str.strip()
```

```python
new_df[new_df["State"] != '-']
```

```python
usa_df = new_df.loc[new_df['Country'] == 'USA']

usa_df
```

```python
usaStates = usa_df['State'].unique()
usaStates
```

```python
usa_df['Location'] = usa_df['State'] + ", " + usa_df['Country']

usa_df.head(2)
```

```python
usaLocations = usa_df['Location'].unique()

# Import Dependicies
import requests
import json

# Import the API key
from config import geoapify_key

baseUrl = "https://api.geoapify.com/v1/geocode/search?text="

stateInfo = []
for location in usaLocations:
    target_url = f"{baseUrl}{location}&format=json&apiKey={geoapify_key}"
    
    geo_data = requests.get(target_url).json()


    stateInfo.append({
        "location": geo_data["query"]["text"],
        "lat": geo_data["results"][0]["lat"],
        "lon": geo_data["results"][0]["lon"]
    })
```

```python
print(stateInfo)
```

```python
with open("location.json", "w", encoding='utf-8') as outfile:
    json.dump(stateInfo, outfile, ensure_ascii=False, indent=4)
```
